# 정규표현식

```python
import re
```

```python
text = 'I love you. \t사랑해.'

eng = "[^A-Za-z ]"
kor = "[^가-힣ㄱ-ㅎㅏ-ㅢ ]"

r1 = re.sub(eng,'',text)
r2 = re.sub(kor,'',text)
```

```python
print(r1)
print(r2.strip())
```

```python
text = '저의 이메일 주소는 alsgur6576@naver.com입니다'
sub_word = '<E-mail>'

p= re.compile('[a-zA-Z0-9\-_.]+@[a-zA-Z0-9\-_.]+')

r = re.sub(p,sub_word,text)
print(r)
```

```python
# 특정 문자열(괄호)의 제거
text = '명지대학교 경영대(경영정보학과, 경영학과, 국제통상학과)의 취업률이 예년 대비 3.5% 상승하였습니다'

p = re.compile('\(+[가-힣0-9, ]+\)')

r = re.sub(p,'',text)
print(r)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
import time
import urllib
```

```python
driver = webdriver.Chrome()
url = 'http://naver.com'
driver.get(url)

driver.find_element_by_class_name('input_text').send_keys('명지대 mcc')
driver.find_element_by_class_name('input_text').send_keys(Keys.ENTER)
driver.find_element_by_css_selector('#lnb > div.lnb_group > div > ul > li:nth-child(2) > a').click()

time.sleep(3)
driver.find_elements_by_class_name('_image._listImage')[0].click()

img = driver.find_element_by_class_name('_image')
src = img.get_attribute('src')

# download the image
urllib.request.urlretrieve(src, "./images/mcc.jpg")
```

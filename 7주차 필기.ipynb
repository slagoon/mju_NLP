# 정규표현식

```python
data = '''
park 800905-1049118
kim 700905-1059119
'''
```

```python
result = []
for line in data.split('\n'):
    word_result=[]
    for word in line.split(' '):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] + '-' + '*******'
        word_result.append(word)
    result.append(' '.join(word_result))
print('\n'.join(result))
```

```python
import re
```

```python
pat = re.compile("(\d{6})[-](\d{7})")
print(pat.sub('\g<1>-*******',data))
```

### 다른 반복표현 : {n},{m,n}
- {n}: 앞 문자가 n번 반복되는 패턴
- {m,n} : 앞문자가 m번 반복되는 패턴부터 n번 반복되는 패턴까지

```python
# {n} 사용 예 - 잘 안쓰임
pattern = re.compile('AD{2}A')
```

```python
pattern.search('ADA')
```

```python
pattern.search('ADDA')
```

```python
pattern.search('ADDDDA')
```

### match와 search
- match : 문자열 처음부터 정규식과 매칭되는 패턴을 찾아서 리턴
- search : 문자열 전체를 검색해서 정규식과 매칭되는 패턴을 찾아서 리턴

```python
pattern = re.compile('[a-z]+')
```

```python
matched = pattern.match('Dave')
```

```python
searched = pattern.search("Dave")
```

```python
print(matched)
print(searched)
```

## findall 함수 : 정규표현식과 매칭되는 모든 문자열을 리스트 객체로 리턴

```python
pattern = re.compile('[a-z]+')
findalled = pattern.findall('Game of Life in Python')
```

```python
print(findalled)
```

```python
pattern2 = re.compile('[A-Za-z]+')
findalled2 = pattern2.findall('Game of Life in python')
print(findalled2)
```

```python
pattern = re.compile('[a-z]+')
findalled = pattern.findall('GAME')

if len(findalled)>0:
    print('정규 표현식에 맞는 문자열이 존재')
else:
    print('정규 표현식에 맞는 문자열이 존재하지 않음')
```

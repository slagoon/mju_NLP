```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.concat([df_train['Fare'], df_train['Embarked']], axis=1)# 운임과 탑승항구 데이터 연결
f, ax = plt.subplots(figsize = (8,6)) # 한번에 여러 그래프를 보여줌
fig = sns.boxplot(x='Embarked', y = 'Fare', data=data)
```

```python
f, ax = plt.subplots(figsize = (8,6))
sns.boxplot(x='Embarked',y='Fare', hue='Pclass',
            data=df_train, palette='Set3')
```
